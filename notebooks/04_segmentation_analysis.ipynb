# 🧩 04: Segmentation Analysis

Deep-dive into the `HDBSCAN` clusters and RFM persona assignments.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import os

sys.path.append(os.path.dirname(os.getcwd()))

from components.C_segmentation.rfm import compute_rfm
from components.C_segmentation.clustering import SegmentationEngine
from data.loader import load_transactions

df = load_transactions()
rfm_df = compute_rfm(df)
engine = SegmentationEngine()
clusters = engine.fit_predict(rfm_df)
rfm_df['cluster'] = clusters
```

## 1. Cluster Sizes
Distribution of merchants across discovered segments.

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=rfm_df, x='cluster', palette='cool')
plt.title("Merchant Count per HDBSCAN Cluster")
plt.show()
```

## 2. RFM Feature Distributions by Cluster
Understanding what defines each segment.

```python
plt.figure(figsize=(14, 8))
for i, col in enumerate(['recency', 'frequency', 'monetary']):
    plt.subplot(3, 1, i+1)
    sns.boxplot(data=rfm_df, x='cluster', y=col)
    plt.yscale('log')
    plt.title(f"{col.capitalize()} by Cluster")
plt.tight_layout()
plt.show()
```
