# 🔍 01: Data and Feature EDA

This notebook explores the raw transaction snapshot data, focusing on distributions, nulls, and potential features for ML models.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import os

# Add project root to path
sys.path.append(os.path.dirname(os.getcwd()))

from data.loader import load_transactions
import config

sns.set_theme(style="whitegrid")
df = load_transactions()
print(f"Dataset shape: {df.shape}")
```

## 1. Null Analysis
Visualizing the sparse fields in the payment schema.

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cbar=False, yticklabels=False, cmap='viridis')
plt.title("Missing Value Heatmap")
plt.show()
```

## 2. Transaction Amount Distribution
Highly skewed, typical for financial data.

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['transaction_amount'], bins=50, kde=True, log_scale=True)
plt.title("Distribution of Transaction Amounts (Log Scale)")
plt.show()
```

## 3. Market Share by Terminal Owner
Identifying the top agents.

```python
top_merchants = df['terminal_owner'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_merchants.values, y=top_merchants.index, palette='magma')
plt.title("Top 10 Terminal Owners by Transaction Count")
plt.show()
```
